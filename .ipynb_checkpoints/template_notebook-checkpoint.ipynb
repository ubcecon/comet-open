# Part 1: Loading Data

In this tutorial, we will be working with the Penn World Table 10.0.  This data is via:

Let's get started by importing the packages and data into our notebook, so that we can use it.  You should also check out the documentation on the link above, if you're not sure what a variable does or represents.

````r
#import packages
library(tidyverse)
library(haven)

#load the data
pwt_data <- read_dta("pwt100.dta")

#declare factors
pwt_data <- as_factor(pwt_data)

pwt_data <- pwt_data %>%
    mutate(countrycode = as.factor(countrycode)) %>%
    mutate(country = as.factor(country)) %>%
    mutate(currency_unit = as.factor(currency_unit))

#check that it looks OK
glimpse(pwt_data)
#there will be a lot of missing data

````

# Part 2:  Graphics in R

We will be using visualizations extensively in this course.  One of the strongest features of R is the ability to quickly and easily create high-quality graphics.  However, it can be a little intimidating at first because a completed graphical command can look very complicated.

This is because R uses a "language" for how graphics are created (called the _grammar of graphics_), which is based on best practices from statistical visualization theory, and centres _data_ in the process.  You can find an excellent guide [as part of this collection of cheatsheets](https://www.rstudio.com/resources/cheatsheets/): this is all part of the ```ggplot``` library, which is included in the ```tidyverse``` package.

 * You should download the ```ggplot``` cheatsheet now; it will be invaluable as a reference guide in these notes, and more generally.

Essentially, in an R graphics, you create a series of "layers" which implement a specific visual output:

* First, identify a dataset from which you want to create your graph (```data=```)
* Then, you associated variables in that dataset to _aesthetics_ (```aes=```)
  * _Aesthetics_ represent different properties of a graph, like "what goes on the $x$-axis", or "what does the color of the line represent".  Each type of visualization is associated with a collection of necessary and optional aesthetic feature; some visualization are incompatible with certain features.
* You can then attach a coordinate system and a plot to the graph (```geom```), which takes the aesthetics and describes them
  * This will also include options like ```position``` which indicates how to combine elements (e.g. stack the bars, or place them side-by-side
* Finally, you can tweak the visualization - such as adding labels or changing the colour scheme

Let's break down a complex example of bivariate chart to see how this works:

````r
NA_data <- filter(pwt_data, (countrycode == "CAN")|(countrycode == "USA")|(countrycode == "MEX"))
#this is the pwt for just north america

plot <- ggplot(data = NA_data,  #this declares the data for the chart; all variable names are in this data
                aes(# this is a list of the aesthetic features of the chart
                    x = year,   #for example, the x-axis will be "year"
                    y = rgdpe/pop, #the y-axes will be expenditure-based real GDP per capita
                    fill = country, #the color fill will be country
                    color = country #so will the outline
                ),
                ) 
    
plot <- plot + labs(x = "Year", #add aesthetic labels
                    y = "Real GDP per capita (expenditure-based)",
                    color = "Country",
                    fill = "Country",
                    title = "North American Real GDP per Capita over Time") #...and titles

plot <- plot + scale_fill_brewer(palette="Accent") #set the colour palette for fills
plot <- plot + scale_color_brewer(palette="Accent") #set the colour palette for outlines



plot1 <- plot + geom_col(     #now we add the visualization (a bar graph)
                            position = "dodge") #this places the visualizations side-by-side

                            #if you change position to "stack" it will be a stacked graph!
                
plot1 #show the plot
````

Beautiful!  It's also very easy for use to quickly change the visualization.  For instance, if we wanted to make this a line graph instead of a bar chart, we could do the following:

````r
plot2 <- plot + geom_line()

plot2
````

Nice!  Okay, to be clear **you need to understand all of that at once**.  Building a visualization like that is the outcome of a process of incremental tweaks and adjustments.  Let's explore this by building a version of this on our own, together.

## Part 3: Building a Chart

Let's start simple, by trying to make something interesting.  How about **"What's the relationship between GDP per capita and human capital in the world today?"**

The first thing we want to do is identify what data we need:
* We will need GDP, which we can get using either ```rdgpe``` or ```rdgpo``` (Quiz: what's the difference?)
* We will need population, which we can get using ```pop```
* We will need human capital, which is the variable ```hc```
* We will need data from today, which is ```year == 2019```, the most recent data in our sample

OK, so let's start out by ```filter```-ing the data to just get 2019 data.

````r
figure_data <- filter(pwt_data, year == 2019)

head(figure_data$year)
````

Nice!  Okay, next we need to start creating our object - however, we should pause for a moment an consider what kind of vizualization we want.  

* Our question is about the relationship between two quantitative variables
* We are interested in understanding how they move together

While there are a couple of options, let's decide that we want to start with a **scatterplot**.  If we consult our cheat-sheet, we can see that scatterplots are the ```geom_point()``` command.  This _requires_ the aesthetic properties:
* ```x```, the $x$-axis
* ```y```, the $y$-axis

We then have a bunch of optional ones, like ```alpha, color, fill, shape, size, stroke``` which we don't have to use.

Let's start simple, and make ```x``` represent human capital, and ```y``` represent real GDP per capita.  We can start our visualization by creating our ```ggplot``` object and assigning all these properties:

````r
figure <- ggplot(data = figure_data, #associate the data we chose
                 aes(
                     x = hc,  #x is human capital
                     y = rgdpe/pop #y is rgdp/pop
                 ))

# note: you can set aethestics to be simple functions of variables!
````

So far, nothing is happening; this is because we need to assign our visualization!  Right now, it's just data and properties.  Let's test it our by adding our ```geom_point()``` layer:

````r
figure + geom_point()
````

Nice!  Already looking pretty good.  But let's make it a little more interesting.  How about if made the size of each point relative to the population?  That way, bigger countries would be more prominent on the graph.  We can do this by assigning the aesthetic again:

````r
figure + geom_point(aes(
                    size = pop))
````

Notice an important fact: you can assign aesthetics on _either_ the ```ggplot``` layer _or_ on a ```geom```.  The difference is that the ```ggplot``` aesthetics are automatically inherited by all other layers; there's no other real difference.

* **Important Note**: generally, any aesthetic property which can be assigned in ```aes()``` can also be assigned to the ```geom``` directly.  For example, if you wanted to make a line dashed or a point red, you could do this by setting ```geom_point(color = "red")```.  However, this will apply to all parts of the ```geom``` so use it wisely!

Now, let's make it more colourful.  How about if we make colour of each point change as the employment (```emp```) rate changes?   So, darker colors would represent higher labour force utilization.  Again, we can do this by assigning the aesthetic:

````r
figure <- figure + geom_point(aes(
                    size = pop,
                    colour = 100*emp/pop))

figure
````

Great work!  We now have a solid visualization.  However, I think that colour looks ugly.  Let's choose a different one.  The easiest way to set colours in R is using palettes.  The list of all the palettes options are:

````r
RColorBrewer::display.brewer.all() 
````

Let's choose ```YlOrRed```.  We can apply this using the following (somewhat cryptic) command:

````r
figure <- figure + scale_color_distiller(palette="YlOrRd")

figure

#you might notice we used "color_brewer" earlier and "color_distiller" here
#one is for discrete visualization (brewer) while the other is for continuous (distiller)
#it doesn't really matter why this is, but watch out for it; R will complain
````

Nice!  However, it doesn't look quite right.  Let's make it look nicer, by adding good labels.  We can do this with the ```labs()``` layer.  Like the ```geom``` layer, it adds information.  We just add text to associate with each aesthetic, and other properties (like a title!):

````r
figure <- figure + labs(
            x = "Human Capital",  #let's give the x aesthetic this label
            y = "Real GDP per capita (expenditure-based)",
            color = "Employment Rate (%)",
            size = "Population (millions)",
            title = "Real GDP per Capita versus Human Capital (2019)"
)

figure
````

Good work!  We now have a publication quality visualization, ready to go.  This is a good example of the process we go through when we build visualizations: a lot of experimenting, a lot of trial and error.  Always ask yourself: "is this effective? Is this what I want it to do?"

R makes it easy to create high-quality, impactful graphics.  Let's try it on our own, next.

## Part 4: Gallery of Options

#### Example 1: Multiple Layers

You can make multiple charts on top of one another by overlaying multiple ```geom```-s.

* It's a good idea to set the aesthetics that are not common per ```geom``` not in the ```ggplot``` object.



````r
eg1 <- ggplot(data = filter(NA_data, countrycode == "USA"), #just the USA, for this example
              aes(
                  x = year #only setting the x aesthetic
              ))

eg1 <- eg1 + geom_line(aes(
                        y = rgdpe/pop,  #now I set the y to be per capita real GDP
                    ),
                       color = "black" #you can set colors globally, not as an aesthetic
                      )

eg1 <- eg1 + geom_col(aes(
                        y = hc*10000, #scale (see below)
                        ), 
                        alpha = 0.1, #make it transparent
                        fill = "red" #make it red
                        )

#to make this look good, you will need a similar scale for the two
#an alternative is to use a second y axis

eg1 <- eg1 + scale_y_continuous( #see a later example for axes
            name = "Real GDP per capital (expenditure-based)", #name of the main axis
            
            sec.axis = sec_axis(trans = ~.*0.0001, name = "Human Capital")
                    #the trans argument is the scale transofrmation for HC
                    #this one muliplies it by 1000
            )

eg1 <- eg1 + labs(title = "Example 1: Overlays")

eg1
````

#### Example 2: Facets

Faceting a graph basically replicates the visualization by a variable in the data; this is called a _facet_ and is a simple way to make comparisons.

````r
eg2 <- ggplot(data = NA_data, aes(
                x = year,
                y = rgdpe/pop
            ))

eg2 <- eg2 + geom_line()

eg2 <- eg2 + labs(x = "Year", y = "Real GDP per capita (expenditure-based)", title = "Facet Example")

eg2 <- eg2 + facet_grid(cols = vars(country))

#facets can be either rows or cols (try changing the "cols")

eg2
````

#### Example 3: Positions

You can change how combinations of aesthetics look using the ```position``` option:

* ```stack``` stacks the elements (like a stacked bar graph)
* ```dodge``` offsets them, so they don't overlap
* ```jitter``` slightly moves them; good for scatterplots that overlaps a lot
* ```nudge``` move labels away from points
* ```fill``` is like stack, but to 100%

Let's look at these options using a single visualization.

````r

eg3 <- ggplot(data = NA_data,  
                aes(
                    x = year,   
                    y = rgdpe, 
                    fill = country, 
                    color = country
                ),
                ) 
    
eg3 <- eg3 + labs(x = "Year", 
                    y = "Real GDP (expenditure-based)",
                    color = "Country",
                    fill = "Country",
                    title = "North American Real GDP over Time") 

eg3 <- eg3 + scale_fill_brewer(palette="Accent") #set the colour palette for fills
eg3 <- eg3 + scale_color_brewer(palette="Accent") #set the colour palette for outlines



eg3 + geom_col(position = "dodge") #side by side
eg3 + geom_col(position = "stack") #stacked
eg3 + geom_col(position = "fill") #100% stacked
````

#### Example 4: Adding Labels

In general, labels can be tricky.  Generally, the best way is to use the ```geom_text``` (not, ironically, ```geom_label```) in combination with the some adjustments to the position.
  * Call-outs are added using the ```annotate``` function with specific positions

````r
eg4 <- ggplot(data = figure_data, aes(
                x = hc,
                y = rgdpe/pop,
                ))

eg4 <- eg4 + geom_point(position = "nudge")
eg4 <- eg4 + geom_text(aes(label = countrycode),
                      nudge_x = 0.1, #adjusting position of labels
                      nudge_y = 0.1,
                      size = 2.5, #adusting font size
                      alpha = 0.7 #making slightly see-through
                      )

eg4 <- eg4 + labs(x = "Human Capital", y = "Real GDP per capita (expenditure-based)", title = "Example 4: Labels")

eg4 <- eg4 + annotate("point", x = 4, y = 100000, color = "red", size = 1.5) # a custom point
eg4 <- eg4 + annotate("text", x = 4.1, y = 97000, color = "red", label = "Goal") #text
eg4
````

#### Example 5: Axis Scales

We actually saw this earlier (with a second axis), but in general you can very broadly adjust aesthetic scales and limits.
* You can check the cheatsheet for the details, but basically there's a separate version of this command for (i) continuous aesthetics, (ii) discrete, (iii) dates, (iv) times.
* You can also set up transformation (such as in millions) or functions like $log()$.  Many common function are built in.

Let's look at a couple of options.

````r
eg5 <- ggplot(data = NA_data,  
                aes(
                    x = year,   
                    y = rgdpe/pop, 
                    fill = country, 
                    color = country
                ),
                ) 
    
eg5 <- eg5 + geom_line() 

eg5 <- eg5 + scale_x_continuous( #scale_???_continuous; the ??? refers to the aesthetic (x, in this case)
                    limits = c(2000,2020),  #set the limits
                    name = "Years"
                    )

eg5 <- eg5 + scale_y_continuous(
                limits = c(15000,65000),
                name = "Log Real GDP per capita (expenditure-based)",
                trans = "log" #use log transformation
                )

eg5
````
