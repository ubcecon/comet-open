#### What is the distribution of probability 

When working with data, the **probability** of seeing a particular value of a given variable is often recorded as the _frequency_ or _density_ of that specific value. For instance, let's say we have a dataset with 5,000 observations and a variable called `birthmonth` which records the month of birth of every participant captured in the dataset. If 500 people in the data were born in October, then `birthmonth`=="October" would have a propability of occuring in an observation 10% of the time.
 
#### What is a random variable 

A **random variable** is a variable whose possible values are numerical outcomes of a random phenomenon, such as rolling a dice or surveys done to random people in random locations at random times. There are two types of random variables, discrete and continuous.

- A **discrete random variable** is one which may take on only a finite number of distinct values. Examples of discrete random variables include the options when you roll dice, the number of children in a family, or the number of defective light bulbs in a box of ten. A good example of a discrete random variable of the data used in this notebook is the age group (`agegrp` in the data) of the census participants as it can take only a finite number of distinct values: 17 different categories to be exact. 

- A **continuous random variable** is one which takes an infinite number of possible values. We can commonly think of this variable as being one that is *measured* rather than merely *categorized*. Examples include height, weight, or how much people earn. In the data, we can see that `wages` and `mrkinc` are great examples of continuous random variables. They can take on an infinite number of possible values.

Now that we've learned some key terminology for describing variables and the values they can take on, we can work with distributions.

#### Probability Density Functions (PDFs)
**Probability Density Functions** are also sometimes referred to as PDFs or probability mass functions.

#### 1. Discrete PDF:

>" The probability distribution of a discrete random variable is the list of all possible values of the variable and their probabilities which sum to  1." - Econometrics with R

In other words, a **Probability Density Functions (PDF)** is the probability of occurrence of all the different values of a variable. The probability of occurrence is also called  **density** or **frequency**.

Suppose a random variable X may take k different values, with the probability that $X = x_{i} $ defined to be $ P(X = x_{i}) = p_{i} $. The probabilities $p_{i}$ must satisfy the following:
1. For each i:
$0<p_{i}<1 $ 

2. 
$p_{1} + p_{2} + ... + p_{k} = 1 $

There are two different ways to view the PDF of a discrete variable. The first way is by creating a density table. 

```r
sample_size <- nrow(census_data) #this is a new command!
table <- census_data %>% 
    group_by(agegrp) %>%
    summarize(Count = n(),
              Frequency = n()/sample_size*100) #creates two variables in our table

table
```

A second way to visualize a pdf is by creating graph. Because a pdf has a finite number of distinct values of which we measure their frequency, the typical format to express a categorical array such as a pdf is through a bar graph following what we learned in _Introduction to Visualization_.

```r
plot <- ggplot(data = table,  #this declares the data for the chart; all variable names are in this data
                aes(# this is a list of the aesthetic features of the chart
                    x = agegrp,   #for example, the x-axis will be "year"
                    y = Frequency, #the y-axes will be expenditure-based real GDP per capita
                ),
                ) 
plot1 <- plot + geom_col()
plot1
```

#### 2. Continuous PDF:

>"Since a continuous random variable takes on a continuum of possible values, we cannot use the concept of a probability distribution as used for discrete random variables." - Econometrics with r


Unlike a discrete variable, a continuous random variable is not defined in specific values. Instead, it is defined over intervals of values, and is represented by the area under a curve (in advanced mathematics, this is known as an integral). The curve, which represents the probability function is also called a density curve and it must satisfy the following:
1. The curve has no negative values p(x) > 0 for all x (there cannot be a negative probability of observing a value)
2. The total area under the curve is equal to 1

Let's imagine a random variable that can take any value over an interval of real numbers. The probability of observing values between a and b is the area between the density curve and the area between a and b:

$\mathrm{P}(a \le X \le b) = \left(\int_{a}^{b} f(x) \; dx\right)$

Since the number of values which may be assumed by the random variable is infinite, the probability of observing any single value is equal to 0. Therefore, to visualize a continuous PDF we commonly use graphs rather than tables and do not use bar graphs or scatterplots to visualize a continuous PDF. Because it is a continuous variable, we need to use a visualization type that allows for the entire continuum of possible values to be represented in the graph. As stated above, the probability of observing values between a and b is the area underneath the curve, therefore a continuous PDF should be visualized as a line graph. Suppose we would like to visulize a continuous PDF for all wages between 400000 and 500000:

```r
census_data1 <- filter(census_data, between(wages, 400000, 500000))

sample_size <- nrow(census_data1) 

table2 <- census_data1 %>% 
    group_by(wages) %>%
    summarize(Count = n(),  
              PDF = n()/sample_size*100) 

plot1 <- ggplot(data = table2,  
                aes(
                    x = wages,   
                    y = PDF, 
                ),
                ) 
plot1 <- plot1 + geom_line()
plot1
```

#### Cumulative Density Function (CDF)
The **Cumulative Density Distribution** function for both discrete and continuous random variables is the probability that the random variable is less than or equal to a particular value. A CDF can only take values between 0 and 1, this is because the total probability that the random variable is less or equal to the biggest value of the variable is 100% or 1 while the probability that the random variable is less or equal to the smallest value of the variable is 0% or 0. 

We use the sign $f(y)$ to represent the PDF of y and use the sign $F(y)$ to represent the CDF of y.


Mathematically, 
$f_{X}(x)$ denotes the probability density function of X. 
The probability that  X  falls between  a  and  b  where  a < b  is $\mathrm{P}(a \le X \le b) = \left(\int_{a}^{b} f_{X}(x) \; dx\right)$

We know that the entire X variable falls between 2 values if the probability of x falling in between them is 1. Therefore X's CDF curve is 
$\mathrm{P}(−∞ \le X \le ∞) = \left(\int_{−∞}^{∞} f_{X}(x) \; dx\right) = 1$

Below we have the CDF of the continuous variable wages. From the graph below we can tell that most people earn between 0-200000 as the probability of people's wages being less than or equal to 200000 is over 80%.

```r
p <- ecdf(census_data$wages)

#plot CDF
plot(p)
```

#### Joint Probability Distribution

So far, we've looked at distributions for single random variables. However, we can also use distributions to analyze the probability of multiple random variables taking on certain values. This brings us to a special type of distribution, the **joint distribution**. Let's say we have two variables, `X` and `Y`, each of which can take on a series of values simultaneously. The joint distribution gives us the probability of seeing each possible combination of `X` and `Y` values at once. These variables can both be discrete, both be continuous, or can be a mix of the two. The logic of calculating the joint distribution flows identically in all three cases. However, we most commonly see joint distributions representing two discrete random variables. An example with two variables from our dataset, `immstat` and `kol`, will clarify.

```r
# creating a table with the frequency of each combination of immstat and kol
joint_table <- prop.table(rbind(table(census_data$immstat, census_data$kol)))
joint_table <- joint_table[, -5]
joint_table <- joint_table[-4, ]
joint_table <- joint_table[-3, ]
joint_table <- as.data.frame.matrix(joint_table)

joint_table
```

Above, we've created a joint distribution for the two discrete random variables, `immstat` and `kol`. Each entry shows the probability of seeing an observation which fits the corresponding row and column category. As has been emphasized throughout this notebook, these "probabilities" are based on frequencies; here, they represent the frequency of occurrence for each combination of `immstat` and `kol` recorded in the dataset. For instance, about half of the people in this dataset are non-immigrants who know only one of Canada's two official languages (English). One thing you will notice from this table is that the entries all sum up to one. *This is a necessary condition for any joint distribution.* While the sum of entries in each row and each column individually does not need (and almost always will not) sum up to 1, the sum of all entries combined must.

Every joint distribution can be represented by a PDF and CDF, just like single random variables. Because a PDF captures the probability of a random variable taking on certain values, for a joint distribution it captures the probability of two random variables taking on certain values at the same time. These are exactly the probabilities in the cells of our table up above. The formal notation of a PDF for two jointly distributed random variabels is below.

> $$f(x, y) = Prob (X = x, Y = y)$$

where $f(x, y)$ is the joint probability density that the random variable $X$ takes on a value of $x$ and the random variable $Y$ takes on a value of $y$ simultaneously.

The PDF for jointly distributed random variables is thus quite simple: it captures the joint probability distribution, just in a 3D format. Imagine an x axis dedicated to all of the possible values for one random variable, and a y axis dedicated to all of the possible values for the other random variable. Then the z axis (the vertical axis) captures the probability of seeing every possible x and y combination. This will look like a series of individual points for two discrete random variables such as `immstat` and `kol` above, each of which corresponds to one of the 20 possible combinations of discrete random variables above. When looking at a joint PDF for a discrete and continuous random variable, the 3D will ressemble slices along the values of the discrete variable, while the joint PDF of two continuous random variables will be a completely smooth 3D shape in its entirety. However, these graphs are beyond the scope of Jupyter, and so they are not shown.

The CDF for jointly distributed random variables follows the exact same logic as with single variables as well. It just instead represents the probability of multiple variables taking on values less than those specified all at once. This is quite non-sensical for two discrete random variables such as `immstat` and `kol`, but we can imagine the CDF of two jointly distributed random variables being much more interpretable if both of those variables are continuous (i.e. `wages` and `mrkinc`). The formal notation of a CDF for two jointly distributed random variables is below.

> $$F(x, y) = Prob({X < x}, {Y < y})$$ <br>

where $F(x, y)$ is the joint cumulative probability that the random variable $X$ takes on a value less than or equal to $x$ and the random variable $Y$ takes on a value less than or equal to $y$ simultaneously.

#### Marginal Probability Distribution 

The table above for `immstat` and `kol` gave us the joint probability distribution for these two discrete random variables. However, we can glean more from the table. For instance, if we add up all of the joint probabilities in a row and in a column, we arrive at another concept known as the **marginal distribution**. This is essentially the probability density function for each individual random variable. It gives the probability of one random variable taking on a series of different values, and it is found from utilizing the joint probabilities in our table above. We can represent the marginal probability density function as follows:

> $$f_{x}(x) = \sum_{y} Prob(X = x, Y = y)$$ <br>

where we sum across all possible joint probabilities of $X$ and $Y$ for a given value of $X$. Identical logic holds for the marginal probability density function of the random variable $Y$ as well.

An example will make this more clear. Let's say we wanted the marginal probability distribution function of the `immstat` variable, we would need to find the marginal probability for all possible values of `immstat`. For `immstat` == "non-immigrants", the marginal probability is the sum of all joint probabilities in that corresponding row: 0.5030 + 0.0847 + 0.1836 + 1.4574e-05. Similar logic can be used to find the marginal probability of `immstat` when `immstat` == "immigrants". We can add a row and column to the table to indicate these marginal probabilities.

```r
# calculating the marginal probabilities of kol and immstat
margprobskol <- as.vector(apply(joint_table, MARGIN = 2, FUN = sum))
margprobsimmstat <- as.vector(apply(joint_table, MARGIN = 1, FUN = sum))

# adding a row and column for the marginal probabilities of kol and immstat
joint_table <- cbind(joint_table, margprobsimmstat)
joint_table <- rbind(joint_table, margprobskol)
rownames(joint_table) <- c("non-immigrants", "immigrants", "margprobskol")
joint_table[3, 5] <- NA

joint_table
```

From the above, we can see the marginal probability distributions for `immstat` and `kol`. Notice that, like the joint probabilities in the interior of the table, the marginal probabilities of both `kol` and `immstat` must individually sum up to 1. This is because the marginal probability distribution for a variable covers the full extent of values that random variable can take on; thus, the individual probabilities of that random variable taking on each possible value (2 for `immstat`, 4 for `kol`) must add up to 1. We can also graph the marginal probability density function for this variable. This will look like the 2D probability density functions we've seen before, since we are now again dealing with a single random variable. The only difference in this case is that we had to find this distribution differently since this variable was jointly distributed with another variable. However, the process for graphing the distribution function is identical to what we have already learned. We can also graph the marginal cumulative distribution function as well, although nothing new would be gleaned from this function: the logic underpinning this function is identical to that of the case with a single random variable and no joint distribution.

#### Conditional Probability Distribution 

Finally, we may want to drill down to one specific value for a given random variable before looking at the probability distribution of another variable. This brings us to the **conditional distribution**. The conditional distribution function gives us the probability of seeing a host of values for one random variable conditional on a specified value of another random variable. In this way, conditional distribution functions allow us to hone in on the probability distribution of a variable under the specific setting that another jointly distributed variable takes on one specific value. There is no easy way to add rows/columns to our table in order to display conditional probabilities; however, we can think about conditional probabilities mathematically. Let's say we want to find the conditional probability of a subject speaking only English given that they are an immigrant. While it may be tempting to locate the cell where these are both the case (`kol` == "english only" and `immstat` == "immigrant"), this is wrong. Remember, these cells represent joint probabilities, the probability of seeing a possible combination of values for our two random variables simultaneously among all possible combinations of values these two variables can take on. With conditional probabilities, we are suddenly looking at a subset of all of the possible combinations, namely the combinations corresponding only to the immigrant row of our table. Thus, we must find the conditional probability differently. We will use the following formula to find the conditional probability density function:

> $$f(x | y) = \frac {Prob ((X = x) \bigcap (Y = y))} {Prob(Y = y)}$$

where $f(x | y)$ represents the conditional probability that the random variable $X$ will take on a value of $x$ when the random variable of $Y$ takes on a value of $y$. The $\bigcap$ symbol simply represents the case that both $X$ = $x$ and $Y$ = $y$ simultaneously (a joint probability). One caveat to this equation is that the marginal probability that $Y = y$ must not be 0, else this conditional probability will be undefined.

In our example, $x$ is the random variable of `kol`, while $y$ is the random variable we are conditioning on: `immstat`. The numerator represents the probability that a person is both an immigrant and speaks only English (a joint probability directly from our table), while the denominator is the marginal probability that a random person is an immigrant (from the final column of our table). This allows us to find the conditional probability that someone speaks only English (among Canada's official languages) given that they are an immigrant: 0.183 / 0.227 ~ 0.806. Thus, conditional on a person having immigrant status in this dataset, there is about an 81% chance that they only speak English among Canada's official languages. It thus follows that our conditional probability distribution as a whole represents the probability of seeing any value of $x$ conditional on any value of $y$. In this way, conditional probability density functions are 3D, much like probability density functions in the case where we aren't dealing with jointly distributed variables at all. For this reason, a graph of a conditional probability density function cannot be shown. However, you can imagine a similar intuition to what we covered for the PDF and CDF of joint probability distributions earlier. Finally, we must remember that a cumulative distribution function exists for this conditional distribution, as is always true for distributions. However, discussing it at length is not necessary, so we merely mention it in passing as was done for marginal cumulative distribution functions for the sake of brevity.

One important point to consider is that of **statistical independence of random variables**. Two random variables are independent if and only if their joint probability of occurrence equals the product of their marginal probabilities for all possible combinations of values of the random variables. In mathematical notation, this means that two random variables are statistically independent if and only if:

> $$f(x, y) = f_{x}(x) f_{y}(y)$$

We can check for statistical independence of our jointly distributed random variables, `immstat` and `kol`, quite quickly by consulting our table up above. Try to deduce whether these variables are independent yourself by multiplying combinations of marginal probabilities to see if they match the joint probability in the corresponding cell.

Throughout these explanations, we have continually referred to the joint, marginal and conditional distribution of two discrete random variables; however, ***one or both of these variables can be continuous***. We focused on discrete random variables since they are much easier to represent in table format. Just imagine trying to create a table of joint probabilities for two jointly distributed continuous random variables: there would be near infinite cells, each of which would have a joint probability of about 0! Nonetheless, all of the same logic applies to continuous random variables. The only difference is that we often refer to mathematical formulas when finding the marginal and conditional probabily functions for these random variables, since their PDFs and CDFs can be represented by mathematical functions (provided they fit the mathematical conditions of a PDF we outlined earlier).

One last important point to remember is that ***we can have more than two jointly distributed random variables***. We may want to represent the probability of 3, 4, or an even larger number of variables taking on certain values at once. While this is possible, it is not easy to represent graphically or in table format. That is why we have stuck to investigating two jointly distributed random variables in this notebook. However, you should keep this in mind as you progress in this course.

#### 1. Uniform Distribution

A variable has a uniform distribution if all values have the same probability of occurrence. Therefore, the variable's density curve  is a rectangle, with constant height across the interval and 0 height elsewhere. Since the area under the curve must be equal to 1, the length of the interval determines the height of the curve.

First, we will start creating a variable whose distribution is uniform. For this task we will use the function `runif()`. `runif()` generates random numbers which together have a uniform distribution. The command is written as `runif(n, min = , max = )`, max and min provide the interval of which the random variables are picked from. 

```r
example_unif <- runif(10000, min = 10, max = 50)
hist(example_unif, freq = FALSE, xlab = 'x')
```

```r
a <- 10
b <- 50
hist(example_unif, 
     freq = FALSE, 
     xlab = 'x',  
     ylim = c(0, 0.03),
     xlim = c(0,70),
     main = "Uniform distribution for the interval [0,100]")
curve(dunif(x, min = a, max = b), 
      from = 0, to = 70, 
      n = 100000, 
      col = "darkblue", 
      lwd = 2, 
      add = TRUE, 
      yaxt = "n",
      ylab = 'probability')
```

Consider the uniform random variable `example_unif` defined on the interval (10,50). Since the interval has width = 40, the curve has height has to equal to 0.025 over the interval and 0 elsewhere. The probability that X is less than or equal to 25 is the area between 10 and 25, or (25-10)*0.025 = 0.375. 

The `dunif()` function calculates the uniform probability density function for a variable and can also calculate a specific value's density.

```r
x_dunif <- seq(0, 100, by = 1) #creating a variable with a uniform distribution
ex.dunif <- dunif(x_dunif, min = 10, max = 60) #calculating the PDF of the variable x_dunif
plot(ex.dunif, type = "o") #plotting the PDF
```

The punif() function calculates the uniform cumulative distribution function for the set of values.

```r
cdf.unif<- punif(x_dunif,                  # Vector of quantiles
      min = 10,            # Lower limit of the distribution (a)
      max = 60,            # Upper limit of the distribution (b)
      lower.tail = TRUE,  # If TRUE, probabilities are P(X <= x), or P(X > x) otherwise
      log.p = FALSE)      # If TRUE, probabilities are given as log
plot(cdf.unif)
```

The qunif() function calculates, based on the cumulative probability, where a specific value is located in the distribution of density helps us to get the uniform quantile distribution probability values for the data values.

```r
x_qunif <- seq(0, 1, by = 0.01)
y_qunif <- qunif(x_qunif, min = 10, max = 50)    
plot(y_qunif) 
```

#### 2. Normal Distribution
We first saw the normal distribution in the Central Tendency notebook, but we are now better equipped to understand its nuances. The normal distribution is a special type of distribution in that it is symmetrical and bell-shaped. Many random variables in natural and social sciences are normally distributed. Height, birth weight or SAT scores all follow a normal distribution. A normal distribution is characterized by its mean μ  and its standard deviation σ, and it is expressed as N(μ,σ2). Without knowing the mean and the standard deviation one cannot calculate the normal distribution.

Function of the PDF:

$f(x) = \displaystyle \frac{x^{−(x−μ)^{2}/(2σ^{2})}}{σ√(2π)}$

A standard normal distribution is is a special normal distribution, characterized for having a mean equal to zero and a standard deviation equal to 1 (μ=0  and  σ=1). Hence, N(0,1)
 - Standard normal variates are often denoted by  Z
 - Usually, the standard normal PDF is denoted by  ϕ and the standard normal CDF is denoted by  Φ

`rnorm()` - this function is used to generate random numbers whose distribution are normal. It generates an amount of random numbers that is specified by the sample size. 

```r
x <- rnorm(10000, mean=0, sd=1) #rnorm(number of observations created, mean, sd)
hist(x, probability=TRUE) #the command hist creates a histogram using variable x,
xx <- seq(min(x), max(x), length=100)
lines(xx, dnorm(xx, mean=0, sd=1))
```

We have used the random values generated to observe the bell shaped distribution of this random variable. As it can be seen in the graph, the mean is zero (that is where all the values converge) and the standard deviation is one therefore, this is a standard normal PDF. We can change the numbers of mean and sd in the `rnorm()` command to make the distribution not standard as well. 

`pnorm()` - this function can give the entire CDF curve of a normally distributed random *variable* as well as give the probability of a normally distributed random *number* to be less than the value of a given number. 

```r
curve(pnorm(x), 
      xlim = c(-3.5, 3.5), 
      ylab = "Probability", 
      main = "Standard Normal Cumulative Distribution Function")
```

```r
pnorm(27.4, mean=50, sd=20) # gives you the CDF at that specific location
pnorm(27.4, 50, 20)
```

`qnorm()`- this function takes the probability value and gives a number whose cumulative value matches the probability value.
`qnorm()` can also create a percent point function (ppf). A ppf is the inverse of the cumulative distribution function. The CDF of a specific value is the probability of a normally distributed value of a random variable to be less than the value of a **given number**. For the percent point function, we start with that probability and then compute the corresponding **given number** for the cumulative distribution. In other words the answer of qnorm will be the number where the area before it is X% of the sample.

```r
 curve(qnorm(x), 
      xlim = c(0, 1), 
      xlab = "Probability",
      ylab = "x", 
      main = "Percentage PoinT Function")
```

```r
qnorm(0.95, mean=100, sd=15)
```

Finally, the function `dnorm()` gives the height of the probability distribution at each point for a given mean and standard deviation. The height of the pdf curve is the density. This function can also be used to calculate the entire density curve, as observed in the command *lines(xx, dnorm(xx, mean=0, sd=1))*

```r
dnorm(100, mean=100, sd=15)
```

#### 3. Student T-Distribution 

The Student's t-distribution can be observed on random continuous variables and it arises when estimating the sampling distribution of a normally distributed population in situations where the sample size is small and the population's standard deviation is unknown. We will explore the concept of sampling distribution in a later module. 

The t distribution is based on the number of observations and the degrees of freedom. A degree of freedom (M) can be defined as the minimum number of independent coordinates that can be changed without violating any constraint imposed. In the case of the t distribution the degrees of freedom = N-1, with N being the sample size. When  M  is large, the t-distribution begins to look like a standard normal distribution. This approximation between standard normal and t distribution can start being noticed around M≥30.

As with the uniform and normal distribution, to generate random values that together have a t-distribution we add the prefixe r to the name of the distribution, rt().

```r
n <- 100
df <- n - 1
samples <- rt(n, df)
hist(samples,breaks = 20, freq = FALSE)
xx <- seq(min(samples), max(samples), length=100)
lines(xx, dt(xx, df))
```

A very interesting characteristic of the t-distribution that can be observed from this graph is that, although it is bell-shaped and symmetrical as the normal distribution, the t-distribution is not as tall as a normal distribution. That means that the data is more spread out than a normal distribution. This characteristic is explained by the central limit theorem (CLT) and law of large numbers (LLN), concepts that will be explained in a future module.

The function dt() calculates the PDF of a particular variable or the density of a particular value of a variable, depending on the sample size and degrees of freedom. In the examples showed below we will be using the variable ex.tvalues which is a sequence of numbers ranging from -4 to 4 with increments of 0.01. Therefore there are 800 numbers generated which means the degrees of freedom = 799.

```r
ex.tvalues <- seq(- 4, 4, by = 0.01)  #generating a sequence of number 
ex_dt <- dt(ex.tvalues, df = 799) #calculating the PDF
plot(ex_dt)     
```

The pt() function calculates the entire CDF curve of a t-distributed random *variable* as well as give the probability of a t-distributed random *number* to be less that the value of a given number. 

```r
ex_pt <- pt(ex.tvalues, df = 799)   #calculating CDF
plot(ex_pt) 
```

qnorm()-- This function takes the probability value and gives a number whose cumulative value matches the probability value.
qnorm() can also create a percent point function (ppf).

```r
ex.qtvalues <- seq(0, 1, by = 0.01)  #generating a sequence of number 
ex_qt <- qt(ex.qtvalues, df = 99)  #calculating the ppf
plot(ex_qt) #plotting the ppf 
```

Beyond these three common distributions, there are many other types of distributions which relatively accurately fit data. Examples include chi-square distribution or f-distribution. However, in rare cases, we may encounter variables for which we cannot fit a distribution. Perhaps the data is distributed wildly and is not well approximated by any common distribution. In these cases, we describe it as a non-parametrical distribution.
